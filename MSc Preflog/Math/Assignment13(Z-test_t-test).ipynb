```python
import pandas as pd
from scipy import stats
import numpy as np

df = pd.read_csv("/content/Titanic-Dataset.csv")


ages = df['Age'].dropna()


mu_0 = 30

t_stat, p_val_t = stats.ttest_1samp(ages, mu_0)

sample_mean = np.mean(ages)
n = len(ages)
sigma = 14

z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))
p_val_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))


print("Sample size (n):", n)
print("Sample mean age:", round(sample_mean, 2))

print("\n--- T-test Results ---")
print("t-statistic:", round(t_stat, 3))
print("p-value:", round(p_val_t, 4))

print("\n--- Z-test Results ---")
print("z-statistic:", round(z_stat, 3))
print("p-value:", round(p_val_z, 4))


alpha = 0.05
print("\nDecision at α = 0.05 (T-test):",
      "Reject H0" if p_val_t < alpha else "Fail to reject H0")
print("Decision at α = 0.05 (Z-test):",
      "Reject H0" if p_val_z < alpha else "Fail to reject H0")

```
