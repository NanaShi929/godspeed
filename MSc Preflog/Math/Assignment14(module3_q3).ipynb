```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
data = iris.data
feature_names = iris.feature_names


X = data[:, :2]


scaling_matrix = np.array([[2, 0],
                           [0, 0.5]])
X_scaled = X.dot(scaling_matrix.T)


theta = np.radians(45)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta),  np.cos(theta)]])
X_rotated = X.dot(rotation_matrix.T)

proj_vector = np.array([1, 1]) / np.sqrt(2)
X_projected_scalar = X.dot(proj_vector)
X_projected = np.outer(X_projected_scalar, proj_vector)


plt.figure(figsize=(12, 10))


plt.subplot(2, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c='blue', alpha=0.6)
plt.title('Original Dataset (Sepal Length vs Sepal Width)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.axis('equal')


plt.subplot(2, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c='green', alpha=0.6)
plt.title('Scaled Dataset (x2 stretch on Sepal Length, x0.5 compress on Sepal Width)')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.axis('equal')


plt.subplot(2, 2, 3)
plt.scatter(X_rotated[:, 0], X_rotated[:, 1], c='red', alpha=0.6)
plt.title('Rotated Dataset (45 degrees)')
plt.xlabel('X\'')
plt.ylabel('Y\'')
plt.axis('equal')


plt.subplot(2, 2, 4)
plt.scatter(X_projected[:, 0], X_projected[:, 1], c='purple', alpha=0.6)
plt.title('Projected Dataset (onto 45 axis)')
plt.xlabel('Projection axis X')
plt.ylabel('Projection axis Y')
plt.axis('equal')

plt.tight_layout()
plt.show()


```
