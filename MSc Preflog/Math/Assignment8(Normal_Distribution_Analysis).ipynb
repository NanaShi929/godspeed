```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


np.random.seed(42)
mean, std, n = 100, 15, 1000
iq_scores = np.random.normal(mean, std, n)


plt.figure(figsize=(8, 5))
count, bins, _ = plt.hist(iq_scores, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
pdf = norm.pdf(bins, mean, std)
plt.plot(bins, pdf, 'r-', linewidth=2, label="Theoretical PDF")
plt.title("Histogram of IQ Scores with PDF")
plt.xlabel("IQ")
plt.ylabel("Density")
plt.legend()
plt.show()


sample_mean = np.mean(iq_scores)
sample_median = np.median(iq_scores)
sample_std = np.std(iq_scores, ddof=1)

print("Sample Statistics:")
print(f"Mean: {sample_mean:.2f} (Theoretical: {mean})")
print(f"Median: {sample_median:.2f}")
print(f"Std Dev: {sample_std:.2f} (Theoretical: {std})")


p_less_90 = norm.cdf(90, mean, std)
p_greater_110 = 1 - norm.cdf(110, mean, std)
p_between_95_105 = norm.cdf(105, mean, std) - norm.cdf(95, mean, std)

print("\nProbability Calculations:")
print(f"P(IQ < 90): {p_less_90:.4f}")
print(f"P(IQ > 110): {p_greater_110:.4f}")
print(f"P(95 < IQ < 105): {p_between_95_105:.4f}")


x = np.linspace(60, 140, 500)
cdf = norm.cdf(x, mean, std)

plt.figure(figsize=(8, 5))
plt.plot(x, cdf, 'b-', linewidth=2, label="CDF")
plt.title("Cumulative Distribution Function (CDF) of IQ")
plt.xlabel("IQ")
plt.ylabel("Cumulative Probability")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
